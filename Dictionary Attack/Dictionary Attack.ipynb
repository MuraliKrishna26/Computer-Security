## Plain Text to Hard Password

Rules :
(a) Every even index character will be replaced into a special character. For this hashmap is used.
(b) Every alternative odd index will be capital letter and non alternative will have small letter.

```python
# Preparing HashMap

map = {
      'a':'@',
      'b':'6',
       'c':'(',
       'd':'P',
       'e':'3',
       'f':'{',
       'g':'?',
       'h':'#',
       'i':'!',
       'j':']',
       'k':'<',
       'l':'1',
       'm':'w',
       'n':'u',
       'o':'0',
       'p':'%',
       'q':'9',
       'r':'&',
       's':'$',
       't':'+',
       'u':'=',
       'v':'^',
       'w':'8',
       'x':'*',
       'y':'4',
       'z':'~',
}
```

```python
def englishToComplex(string) :
    ans = ""
    j = 0
    for i in range(len(string)):
        if i % 2 == 1 :
            ans += map[string[i].lower()]
        else :
            if j == 0 :
                ans += string[i].lower()
                j = j + 1
            else :
                ans += string[i].upper()
                j = j - 1     
    return ans  
```

```python
plainEnglish = input("Enter Plain English Text : ")
changedPassword = englishToComplex(plainEnglish)
print("Encrypted Text : ", changedPassword)
```

```python
def getCrackedPassword(length):
    
    word_with_len = []
    
    for word in lines:
        if len(word) == length :
            hardpassword = englishToComplex(word)
            word_with_len.append(hardpassword)
    
    print("Enter Prefered Choice : ")
    print(" 1. Print HardPassword ")
    print(" 2. Write in a File ")
    
    k = int(input(""))
        
    if k == 1 :
        
        print("Number of possible passwords are : ",len(word_with_len))
        print(word_with_len)
        
    elif k == 2 :
        
        text_file = open("output.txt","w")
        
        string = " All the posibble strong passwords with length " + str(length) + " are : \n"
        
        for i in word_with_len :
            string += i
            string = string + " \n"
        
        
        text_file.write(string)
        
        print("Output file is updated")
        
    else:
        print("Invalid Input")
        

length = int(input("Enter the length of the password "))
getCrackedPassword(length)        
```
